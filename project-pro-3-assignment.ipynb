```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import to_date , col
```

```python
if __name__ == "__main__":

    spark = SparkSession.builder\
            .appName("ETL and LET")\
            .master("local[*]")\
            .getOrCreate()
    
    mydf = spark.read.format("csv")\
            .option("header","true")\
            .load("file:///home/saif/cohort_F11/datasets/datasets_project_pro/airlines1.csv")

```

```python
#mydf.show(truncate=False)
#mydf.printSchema()
```

```python
mydf.write.format("jdbc") \
        .mode("overwrite") \
        .option("url", "jdbc:mysql://localhost:3306/retail_db?useSSL=False") \
        .option("user", "root") \
        .option("password", "Welcome@123") \
        .option("dbtable", "airlines") \
        .option("driver", "com.mysql.cj.jdbc.Driver") \
        .save()
print("Successfull")
```
